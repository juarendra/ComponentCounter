# YOLOv8n → TFLite — SMD Component Detection

Colab usage: set your Roboflow API key in the `API_KEY` variable below, then **Runtime → Run all**. Download the exported `*_float32.tflite` and copy it to `app/src/main/assets/best_float32.tflite`. See `MODEL_IO.md` for the verified runtime contract (416 input, anchor-free `[1,8,3549]`).

```
!pip install -q ultralytics roboflow
from roboflow import Roboflow
from ultralytics import YOLO
API_KEY = ""  # <-- paste your Roboflow API key
ds = Roboflow(api_key=API_KEY).workspace("dainius").project("smdcomponents").version(2).download("yolov8")
model = YOLO("yolov8n.pt")
model.train(data=f"{ds.location}/data.yaml", imgsz=416, epochs=100, batch=16)
print("Exported:", model.export(format="tflite", imgsz=416))
# Download the *_float32.tflite from the runs/ folder and copy to app/src/main/assets/best_float32.tflite
```
